## Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import sklearn as sk
```

## Step 2. Import the dataset

```python
data = pd.read_csv("iphones.csv")
data
```

## Step 3. Import LinearRegression model from linear_model family 

```python
from sklearn.linear_model import LinearRegression
```

## Step 4. Create LinearRegression Model

```python
lr = LinearRegression()
```

## Step 5. Splitting dataset in to input and output

```python
x = data['iphonenumber'].values.reshape(-1,1)
y = data['price']
```

## Step 6. Fit the Model (Training)

```python
lr.fit(x,y)
```

## Step 7. Prdict the value

```python
pridected = lr.predict(x)
lr.predict([[14]])
```

## Step 8. Result Visualization

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(data['iphonenumber'],data['price'],c='r',marker="*")
plt.plot(data['iphonenumber'],pridected,c='b',marker="^")
```

## Dataset - 2 - Salary Dataset

## Step 1. Import the necessary libraries

```python
import pandas as pd
import sklearn as sk
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

## Step 2.  Read the Data

```python
data1 = pd.read_csv("Salary_Data.csv")
```

## Step 3. View Data using Scatter plot 

```python
x=data1['YearsExperience']
y=data1['Salary']

plt.scatter(x,y)
plt.show()
```

## Step 4. Splitting dataset in to input and output

```python
x=data1.iloc[:,:-1].values
y=data1.iloc[:,:1].values
y
```

## Step 5. Splitting dataset in to input and output

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/3,random_state=0)
```

## Step 6. Create LinearRegression Model

```python
lr=LinearRegression()
```

## Step 7. Fit the Model (Training)

```python
lr.fit(x_train,y_train)
```

##  Step 6. Predicting the Test set results

```python
y_pred=lr.predict(x_test)
y_pred
```

##  Step 7. Visualising the Training set results

```python
import matplotlib.pyplot as plt

plt.scatter(x_train,y_train,color='red')
plt.plot(x_train,lr.predict(x_train),color='blue')
plt.title('Salary vs Experience (Training Set')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

##  Step 8. Visualising the Test set results

```python
import matplotlib.pyplot as plt

plt.scatter(x_test,y_test,color='red')
plt.plot(x_test,lr.predict(x_test),color='blue')
plt.title('Salary vs Experience (Training Set')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
